# 결정 트리

## 결정 트리 (Decision Tree, 의사 결정 트리, 의사 결정 나무)

- 분류(classification)와 회귀(regression) 모두 가능한 지도 학습 모델 중 하나
- 스무고개 하듯 예/아니오 질문을 이어가며 학습함


(출처: 텐서 플로우 블로그)


- 이렇게 특정 기준(질문)에 따라 데이터를 구분하는 모델을 결정 트리 모델 이라고 함
- 한 번의 분기마다 변수 영역을 두 개로 나눔

## 가지치기 (Pruning)

- Decision Tree는 과적합(overfitting)이 될 가능성이 높다는 약점 가짐
    - 트리에 가지가 너무 많다면 오버피팅


- 가지치기(pruning) : 나무의 가지를 치는 방법
    - 최대 깊이, 터미널 노드의 최대 개수, 혹은 한 노드가 분할하기 위한 최소 데이터 수를 제한


- min_sample_split을 통해 한 노드에 들어있는 최소 데이터 수 정하기
- max_depth를 통해 최대 깊이 지정해주기

------

- but, 가지치기를 통해서도 충분히 과적합 해결이 x
- 좀 더 일반화 된 트리를 만들어보자
      -> 랜덤 포레스트의 기원이 되는 아이디어
      
      
      
- 랜덤 포레스트가 생성한 일부 트리는 overfitting이 될 수 있지만, 많은 수의 트리를 생성함으로써 overfitting이 예측하는 데에 있어 큰 영향을 미치지 못하도록 예방.

# 랜덤 포레스트

##### 결정트리(Decision Tree)가 모여 랜덤 포레스트(Random Forest)를 구성한다.

- 학습 전용 데이터를 랜덤 샘플링하여 다수의 결정 트리를 만들고, 만들어진 결정 트리들의 결과들을 모아 가장 많이 나온 값을 최종 결과로 도출하는 알고리즘 (다수결의 원칙)
- 이렇게 의견을 통합하거나 여러 가지 결과를 합치는 방식: 앙상블(ensemble)
    - 랜덤포레스트는 ensemble machine learning model
- 여러 개의 작은 결정 트리가 예측한 값들 중 가장 많은 값(분류) or 평균값(회귀)을 최종 예측 값으로 예측하는 것



(출처: 텐서플로우 블로그)
- 오른쪽 아래가 랜덤 포레스트의 decision boundary
- 나머지는 결정 트리 각각의 decision boundary
- 결정 트리의 경계는 다소 모호하고 오버피팅이 되어있지만, 5개의 결정 트리 경계를 평균 내어 만든 랜덤 포레스트의 경계는 보다 깔끔하다.

## 랜덤 포레스트의 장단점

(1) 장점
- 분류, 회귀에서 가장 널리 쓰이는 알고리즘으로 성능이 좋고 정확도 높음
- 간편하고 빠름
- 큰 데이터 셋에서도 잘 사용되며, 많은 입력 변수들을 다룰 수 있음

(2) 단점
- 속도와 메모리의 비용이 상대적으로 큼 (linear에 비해)
    - 더 많은 트리 -> 정확도는 높아지나 but 시간과 리소스 소모가 큼
- 트리 깊이와 개수 설정을 잘못하면 overfitting(과적합) 발생

# 배깅 (Bagging)

- 랜덤포레스트는 제일 먼저 bagging이라는 과정을 거침
- Bagging 방식으로 각자의 데이터를 샘플링해 개별적으로 학습한 후 최종적으로 투표를 통해 가장 보편적인 예측결정을 하게 됨
- Bagging은 트리를 만들 때 training set의 부분집합을 활용해 형성하는 것
ex) training set에 1000개의 데이터가 있을 때, 100개의 데이터만 임의로 선택하여 트리를 만드는데 활용하는 것
- 즉, 모든 트리는 각기 다른 데이터를 바탕으로 형성되지만 모두 training set의 부분집합임

(출처: Medium)


- 이렇게 100개의 데이터를 임의로 선택할 때 중복을 허용한다.

## 랜덤 포레스트

: 훈련을 통해 구성해 놓은 다수의 의사결정트리들로부터 분류 결과를 취합해 예측하는 앙상블 모델


- 랜덤 포레스트 알고리즘은 결정트리(Decision Tree) 분류기 여러 개를 훈련시켜 그 결과들을 가지고 예측을 하는데 각각의 결정 트리를 훈련시킬 때 배깅(Bagging), 즉, 전체 훈련 세트에서 중복을 허용하여 샘플링한 데이터셋을 개별 결정트리 분류기의 훈련 세트로 하여 훈련시키는 방식

- 앙상블 방식을 사용하므로 몇몇 결정트리들이 과대적합 or 과소적합되어도 해당하는 오차가 상대적으로 무시되는 경향이 있음
    - -> 일반적으로 의사결정트리 하나를 사용한 모델보다 더 좋은 성능 가짐



- 즉, 결정트리에 무작위성(Random feature)를 추가한 것

- 랜덤 포레스트 알고리즘은 결정 트리들로 이루어져있으므로 각각의 결정 트리들을 학습하는 과정이 필요함
- 랜덤 포레스트 방식에서는 무작위로 선택한 특성 후보들 중 최적의 특성을 찾는 방법을 사용

### 데이터 가져오기

```python
import pandas as pd

data = pd.read_csv("mushrooms.csv")
data.head()
```

```python
data.isnull().sum()
```

### 데이터 전처리

```python
X = data.drop('class', axis=1) # 학습용 데이터
y = data['class'] # label 데이터

X.head()
```

```python
# 데이터 내부의 기호(알파벳)을 모두 숫자형으로 변환 -> 학습 모델이 처리할 수 있도록
# LabelEncoder 이용

from sklearn.preprocessing import LabelEncoder

Encoder_X = LabelEncoder()
for col in X.columns:
    X[col] = Encoder_X.fit_transform(X[col])

Encoder_y = LabelEncoder()
y = Encoder_y.fit_transform(y)

X.head()
```

```python
y

# 독이 있음(1) / 독이 없음(0)
```

```python
# 범주형(categorical) 변수를 dummy 변수로 나타내기

X = pd.get_dummies(X, columns=X.columns, drop_first=True)
X.head()
```

### 데이터 학습 및 평가

```python
from sklearn.model_selection import train_test_split

# 평가용으로 30%의 데이터 사용
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# (평균이 0, 표준편차가 1인) 표준 정규분표로 표준화
# -> 데이터 노이즈를 줄여 overfitting 확률을 낮출 수 있다.

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#### 주성분 분석 (Principal component analysis, PCA)

- 고차원의 데이터를 저차원의 데이터로 환원시키는 기법

- 데이터를 정사영시켜 차원을 낮춘다면, 어떤 벡터에 데이터들을 정사영시켜야 원래의 데이터 구조를 제일 잘 유지할 수 있을까?
- PCA는 종합점수를 '잘' 계산하는 방법

```python
# 주성분 분석(PCA)를 통해 차원 축소, 낮은 차원의 데이터를 찾아낸다.

from sklearn.decomposition import PCA
pca = PCA(n_components=2)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

### 랜덤 포레스트(Random Forest) 예측

```python
# 랜덤 포레스트 분류기

from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=50, criterion='entropy', random_state=42)
classifier.fit(X_train, y_train)

# 50개의 결정 트리 생성, 엔트로피 분리 기준
```

#### 엔트로피 (Entropy)



** 불순도(Impurity)
- 해당 범주 안에 서로 다른 데이터가 얼마나 섞였는지
- 한 범주에 하나의 데이터만 있다: 불순도 최소
- 한 범주에 서로 다른 두 데이터가 정확히 반반 있다: 불순도 최대
- 결정 트리는 불순도를 최소화 하는 방향으로 학습 진행


** 엔트로피(Entropy)
- 불순도를 수치적으로 나타낸 척도
- 엔트로피가 높음 = 불순도가 높음
- 엔트로피가 1 = 불순도 최대
- Pi : 범주 i에 속한 데이터의 비율

#### 사이킷런이 지원하는 분류 성능평가



1. classification_report (분류 결과표)
- 각각의 클래스를 양성 클래스로 보았을 때의 정밀도, 재현율, f1점수를 구해줌
- 그 평균값으로 전체 모형의 성능을 평가함

2. confusion matrix (오차 행렬)
- 타겟의 원래 클래스와 모형이 예측한 클래스가 일치하는지를 갯수로 센 결과를 표로 나타낸 것

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def print_score(classifier, X_train, y_train, X_test, y_test, train=True):
    if train == True:
        print("Training results:\n")
        print("Accuracy Score: {0:.4f}\n".format(accuracy_score(y_train, classifier.predict(X_train))))
        print("Classification Report:\n{}\n".format(classification_report(y_train, classifier.predict(X_train))))
        print("Confusion Matrix:\n{}\n".format(confusion_matrix(y_train, classifier.predict(X_train))))
        res = cross_val_score(classifier, X_train, y_train, cv=10, n_jobs=-1, scoring="accuracy")
        print("Average Accuracy:\t{0:.4f}\n".format(res.mean()))
        print("Standard Deviation:\t{0:.4f}".format(res.std()))
    elif train == False:
        print("Test results:\n")
        print("Accuracy Score: {0:.4f}\n".format(accuracy_score(y_test, classifier.predict(X_test))))
        print("Classification Report:\n{}\n".format(classification_report(y_test, classifier.predict(X_test))))
        print("Confusion Matrix:\n{}\n".format(confusion_matrix(y_test, classifier.predict(X_test))))
    
```

- 매개변수 값이 학습이라면, 결과 값을 낼 때 10번의 교차검증을 모든 cpu를 사용하여 진행한다.

#### classification report (분류 결과표)

(1) precision(정밀도)
- 양성 클래스에 속한다고 출력한 샘플 중 실제로 양성 클래스에 속하는 샘플의 비율
- TP / (TP+FP)


(2) recall(재현율)
- 실제 양성 클래스에 속한 표본 중 양성 클래스에 속한다고 출력한 표본의 수의 비율
- TP / (TP+FN)


(3) f1점수
- f점수: 정밀도와 재현율의 가중조화평균
- 정밀도에 주어지는 가중치(베타)가 1인 경우, 특별히 f1점수라고 함
- 2*precision*recall/(precision+recall)


(4) accuracy: 정확도
- 전체 학습 데이터의 개수 중 각 클래스에서 자신의 클래스를 정확히 맞춘 개수의 비율


(5) macro : 단순 평균


(6) weighted : 각 클래스에 속하는 표본의 개수로 가중 평균

#### confusion matrix (오차행렬)
- 정답 클래스는 행(row), 예측한 클래스는 열(col)로 나타냄
- TP(양성예측이 맞음) FN(음성예측이 틀림)
- FP(양성예측이 틀림) TN(음성예측이 맞음)

### 랜덤 포레스트(Random Forest) 예측 2

#### 랜덤포레스트 하이퍼 파라미터 튜닝

(1) n_estimators
- 결정 트리의 갯수 지정
- default=10
- 무작정 트리 개수 늘리면 수행 시간이 오래 걸릴 수 있음


(2) min_samples_split
- 노드를 분할하기 위한 최소한의 샘플 데이터 수
    - -> 과적합 제어하는데 사용
- default=2
- 작게 설정할수록 분할 노드가 많아져 과적합 증가


(3) min_samples_leaf
- 리프 노드가 되기 위해 필요한 최소한의 샘플 데이터 수
- 불균형 데이터의 경우 특정 클래스의 데이터가 극도록 작을 수 있으므로 작게 설정 필요


(4) max_features
- 최적의 분할을 위해 고려할 최대 feature 수
- default='auto'
- int형, sqrt 또는 auto, log


(5) max_depth
- 트리의 최대 깊이
- default='None'
- 완벽하게 클래스 값이 결정될 때까지 분할 or 데이터 수가 min_samples_split보다 작아질 때까지 분할
- 깊이가 깊어지면 과적합 위험


(6) max_leaf_nodes
- 리프노드의 최대 개수

```python
from sklearn.model_selection import train_test_split

# 평가용으로 30%의 데이터 사용
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# (평균이 0, 표준편차가 1인) 표준 정규분표로 표준화
# -> 데이터 노이즈를 줄여 overfitting 확률을 낮출 수 있다.

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = RandomForestClassifier()
model
```

#### GridSearchCV를 통한 랜덤포레스트의 하이퍼 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = {'n_estimators': [10, 100], # 결정트리의 개수
        'max_depth': [6, 8, 10, 12], # 트리의 최대 깊이
        'min_samples_leaf': [8, 12, 18], # 리프 노드의 최소 샘플 데이터 수
        'min_samples_split': [8, 16, 20]} # 노드 분할하기 위한 최소 샘플 데이터 수


# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rf_clf = RandomForestClassifier(random_state=0, n_jobs=1) # random_state=0: 수행할 때마다 동일한 예측 결과 위해
grid_cv = GridSearchCV(rf_clf, param_grid=params, cv=3, n_jobs=-1)
grid_cv.fit(X_train, y_train)

print("최적 하이퍼 파라미터: ", grid_cv.best_params_)
print("최적 예측 정확도: {:.4f}".format(grid_cv.best_score_))
```

```python
# 위 결과로 나온 쵲거 하이퍼 파라미터를 이용해 다시 모델을 학습하여
# 테스트 데이터에서 예측 성능을 측정

rf_clf2 = RandomForestClassifier(n_estimators=100, 
                                max_depth=8,
                                min_samples_leaf=8,
                                random_state=0,
                                n_jobs=-1)
rf_clf2.fit(X_train, y_train)

pred = rf_clf2.predict(X_test)
print("예측 정확도: {:.4f}".format(accuracy_score(y_test, pred)))
```
